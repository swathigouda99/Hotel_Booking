```python
df['month'] = df['reservation_status_date'].dt.month
sns.countplot(x='month',hue='is_canceled',data=df)
plt.legend(labels=['not_canceled','canceled'])
plt.title('Reservation status per month')
```

In plot we see less reservation cancelation in month of the August and septmber and most cancelation happend in month of Janurary.
